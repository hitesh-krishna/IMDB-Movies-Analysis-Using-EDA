-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here
# n=2 is used so that split can be done only on first 2 portions (as asked in question above)
# Expand True would return DF
genre_all = movies['genres'].str.split('|', n = 2, expand = True) 
movies['genre_1']=genre_all[0]
movies['genre_2']=genre_all[1]
```

```python
#check NULL values in newly added columns genre_1 and genre_2
movies.isnull().sum()
```

```python
#Replace NULL in genre_2 with genre_1 values
movies.loc[pd.isnull(movies['genre_2']),['genre_2']]=movies['genre_1']
```

```python
#check NULL values in newly added columns again
movies.isnull().sum()
```

```python
#Check for the shape of original dataframe
movies.shape
```

```python
# Write your code for grouping the dataframe here

movies_by_segment = movies.groupby(['genre_1','genre_2'])
movies_by_segment
```

```python
# Write your code for getting the 5 most popular combo of genres here
#Note that genre_1 Adventure having 3 genre_2 and clubbed together
PopGenre=movies_by_segment['gross'].mean().sort_values(ascending = False).head(5).to_frame()
PopGenre
```

```python
# Write the code for calculating decade here
# As decade is asked to have 's' in last, its converted to string
movies['decade'] = movies['title_year'].map(lambda x: str(int(x//10)) +'0s')
movies.sort_values(by=['decade'],ascending=False,inplace=True)
movies.head()
```

```python
#Check for the shape of original dataframe
movies.shape
```

```python
# Write your code for creating the data frame df_by_decade here 
df_by_decade=movies.groupby('decade')['num_voted_users'].sum().to_frame().sort_values(['decade','num_voted_users'],ascending =(False,False))
df_by_decade
```

```python
# Write your code for plotting number of voted users vs decade
# Note that in out dataset we have data only halfway of the latest decade hence 2010s votes are lesser
sns.barplot(x='decade', y='num_voted_users', data=movies,estimator=np.sum)
plt.title('--Votes_by_decade--')
plt.show()
```
